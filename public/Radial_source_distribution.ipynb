```python
# at a bunch of radii: calculate the volume, number of sources inside the sphere, give sources / volume
```

```python
stacked = np.concatenate([radNB3,radNB6,radMB3,radMB6])
radii = np.linspace(np.min(stacked), np.max(stacked), 10000)
```

```python
radiiNB3 = np.linspace(np.min(radNB3), np.max(radNB3), 10000)
radiiMB3 = np.linspace(np.min(radMB3), np.max(radMB3), 10000)
radiiNB6 = np.linspace(np.min(radNB6), np.max(radNB6), 10000)
radiiMB6 = np.linspace(np.min(radMB6), np.max(radMB6), 10000)
```

```python
def get_dens(radii, rad):
    radii = (radii).to(u.pc)
    volumes = (radii**3)*4/3*np.pi
    sources_per_volume = []
    for i in range(len(radii)):
        num_sources_within = len(rad[rad.to(u.AU)<=radii[i].to(u.AU)])
        sources_per_volume += [num_sources_within/volumes[i]]
    
    return sources_per_volume/(u.pc)**3
```

```python
densityNB3 = get_dens(radiiNB3, radNB3)
densityNB3[densityNB3 == 0]=np.nan
densityMB3 = get_dens(radiiMB3, radMB3)
densityMB3[densityMB3 == 0]=np.nan
densityNB6 = get_dens(radiiNB6, radNB6)
densityNB6[densityNB6 == 0]=np.nan
densityMB6 = get_dens(radiiMB6, radMB6)
densityMB6[densityMB6 == 0]=np.nan
```
